### We now want to check for wrong datatypes.

```python
# First we want to check which are the columns in our dataframe.
df.columns
```

```python
# For each of these columns we want to check if there are wrong datatypes.
# The first step is to check the dtypes of each column in the dataframe.
df['age'].dtypes
```

```python
# Now that we know the dtypes of each column we can wrote a loop that checks for each column if there are wrong datatypes.
# First step is to create a dictionary mapping the the dtype to the type of an indiviual value in the column.
value_types = {
    'int64' : int,
    'object' : str,
    'float64' : float
}
value_types
```

```python
# We then use a for loop for all the columns to use this mapping to check wrong types.
# We then put the columns in a dictionary with a mapping of a list that says if each column in the dictionary contains a wrong data type.
wrong_data_types = {}
for column in df.columns:
    wrong_data_types[column] = [x for x in df[column] if type(x) != value_types[str(df[column].dtypes)]]

wrong_data_types
```

### No wrong datatypes in any of of the columns.
### This means our data is clean and we can begin analysis.

#### Question 2: What is the average age of men in our database?


```python
# First we look at our dataframe
df
```

```python
# We then isolate the dataframe to include only men.
men_in_dataframe = df[df['sex'] == 'Male']
men_in_dataframe
```

#### Question 3: What is the percentage of people with Bachelor's Degree?

```python
# First of all looking at our dataframe?
df
```

```python
# Isolating the bachelors in education
bachelors = df['education'][df['education'] == 'Bachelors']

print('The percentage of people with Bachelors education is', (len(bachelors) / len(df['education']))*100)
```

#### Question 4: What is the percentage of people with higher education (Bachelor's, Masters or Doctorate) that make >50K?

```python
# Seeing the dataframe
df
```

```python
# Filtering out relevant information
df_filtered = df[(df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])) & (df['salary'] == '>50K')]
```

```python
# Getting the percentage
percentage = (len(df_filtered) / len(df[df['education'].isin(['Bachelors', 'Masters', 'Doctorate'])])) * 100

print('The percentage of people with higher education that earn more than 50K is', percentage)
```
